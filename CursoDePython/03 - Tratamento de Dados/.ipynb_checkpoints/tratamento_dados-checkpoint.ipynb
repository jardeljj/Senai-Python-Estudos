## Modificando os tipos dos dados

```python
df.info()
```

```python
df.head()
```

```python
df['matricula_funcionario'] = df['matricula_funcionario'].astype('str')
```

```python
df = df.astype({
    'codigo_produto': 'str'
})
```

```python
import numpy as np
```

```python
df = df.astype({
    'preco_custo': np.float64,
    'valor_unitario': np.float64,
})
```

```python
float('210.9')
```

```python
df.info()
```

### Alterando strings

```python
df['preco_custo'] = df['preco_custo'].str.replace(',', '.')
```

```python
df['valor_unitario'] = df['valor_unitario'].str.replace(',', '.')
```

```python
df['valor_unitario']
```

### Alterando tipo para Date / Datetime

```python
df['dt_venda'] = pd.to_datetime(df['dt_venda'], format='%d/%m/%Y %H:%M')
df['dt_entrega'] = pd.to_datetime(df['dt_entrega'], format='%d/%m/%Y %H:%M')
```

```python
df.info()
```

## Ordenando dados

```python
dados = {
    'nome': ['Maria', 'João', 'José', 'Ana', 'Gabriel', 'Lucas', 'Ana'],
    'idade': [70, 43, 35, 54, 15, 13, 64],
    'estado': ['MG', 'PB', 'AL', 'RJ', 'BA', 'RS', 'ES']
}

df_dados = pd.DataFrame(dados)
```

```python
df_dados
```

```python
df_dados.sort_values(['nome', 'idade'], ascending=[True, False], inplace=True)
```

```python
df_dados['nome'].sort_values()
```
