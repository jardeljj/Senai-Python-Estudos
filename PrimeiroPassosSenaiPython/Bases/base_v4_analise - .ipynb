```python
#importação de dados
```

```python
# Gráfico de Disperção duplo
fig,(ax1,ax2, ax3) = plt.subplots(1,3, sharey=True, figsize=(12,5))

ax1.scatter(df['Anos_Educ_Superior'],df['Salario'])
#ax2.scatter(df['Tempo_Empresa'],df['Salario'])
ax3.scatter(df['Tempo_de_Carreira'],df['Salario'])
```

```python
# Correlação linear de Pearson

fig = plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), cmap='RdBu_r', vmin=-1, vmax=1, annot=True)
```

```python
# ajuste da técnica de modelo de regressão statsmodels
```

```python
# variavel resposta 

y = df ['Salario']

# Varial explicativa 
df['intercepto'] = 1

x = df[['intercepto','Anos_Educ_Superior','Tempo_de_Carreira']]
```

```python
# Ajuste do melhor modelo
modelo = sm.OLS(y,x)
result = modelo.fit()
```

```python
# Resultado do ajustado

print(result.summary())
```
