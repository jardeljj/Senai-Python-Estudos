```python
#Manipulando dados
import numpy as np
import pandas as pd

#pacotes graficos
import matplotlib.pyplot as plt
import seaborn as sns

#Pacotes de modelagem estatística
import statsmodels.api as sm
```

```python
#importação de dados
```

```python
df = pd.read_csv('base_funcionarios_v2.csv', sep =";", index_col='id')
```

```python
#Analise exploratoria dos dados
```

```python
# Gráfico de Disperção duplo
fig,(ax1,ax2) = plt.subplots(1,2, sharey=False, figsize=(12,5))

ax1.scatter(df['Anos_Educ_Superior'],df['Salario'])
m,b = np.polyfit(df['Anos_Educ_Superior'],df['Salario'],1)
ax1.plot(df['Anos_Educ_Superior'], m*df['Anos_Educ_Superior'] + b)

ax2.scatter(df['Tempo_Empresa'],df['Salario'])
m,b = np.polyfit(df['Tempo_Empresa'],df['Salario'],1)
ax2.plot(df['Tempo_Empresa'], m*df['Tempo_Empresa'] + b)
```

```python
# Correlação linear de Pearson

fig = plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), cmap='RdBu_r', vmin=-1, vmax=1, annot=True)
```

```python
# ajuste da técnica de modelo de regressão statsmodels
```

```python
# variavel resposta 

y = df ['Salario']

# Varial explicativa 
df['intercepto'] = 1

x = df[['intercepto','Anos_Educ_Superior','Tempo_Empresa']]
```

```python
# Ajuste do melhor modelo
```

```python
modelo = sm.OLS(y,x)
result = modelo.fit()
```

```python
# Resultado do ajusto

print(result.summary())
```
