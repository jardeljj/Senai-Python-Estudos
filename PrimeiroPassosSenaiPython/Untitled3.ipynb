```python
colunas = ['Segmento', 'Pais', 'Produto', 'Qtde_de_Unidades_Vendidas',
       'Preco_Unitário', 'Valor_Total', 'Desconto', 'Valor_Total_Desconto',
       'Custo_Total', 'Lucro', 'Data', 'Mes', 'Ano']

df.columns = colunas
```

```python
colunas = ['Valor_Total_Desconto', 'Custo_Total', 'Lucro']
```

```python
df_segmento = df.groupby('Segmento')[colunas].sum().round(2)
```

```python
df_segmento
```

```python
df_segmento['Margem_lucro'] = df_segmento.Lucro / df_segmento.Valor_Total_Desconto
```

```python
df_segmento
```

```python
df_segmento = df_segmento.drop('margem_lucro', axis=1)
```

```python
df_segmento['Margem_lucro'] = (df_segmento.Lucro / df_segmento.Valor_Total_Desconto * 100).round(1)
```

```python
df_segmento
```

```python
#criando graficos com matplotlib
plt.figure(figsize=(15,6))
plt.bar(x=df_segmento.index, height=df_segmento['Margem_lucro'], color = ['red'])
plt.title('Margem de Lucro por Segmento')
plt.xlabel('Segmento')
plt.ylabel('% Margemde de lucro')
```

```python
df_pais = df.groupby('Pais')[colunas].sum().round(2)
```

```python
df_pais
```

```python
df_pais['Margem_lucro'] = df_pais.Lucro / df_pais.Valor_Total_Desconto
df_pais
```

```python
df_pais['Margem_lucro'] = (df_pais.Lucro / df_pais.Valor_Total_Desconto * 100).round(1)
df_pais
```

```python
#criando graficos com matplotlib
plt.figure(figsize=(14,7))
plt.bar(x=df_pais.index, height=df_pais['Margem_lucro'], color = ['purple'])
plt.title('Margem de Lucro por País')
plt.xlabel('País')
plt.ylabel('% Margemde de lucro')
```

```python

df_2019 = df[df.Ano == 2019]
colunas = ['Valor_Total_Desconto', 'Custo_Total', 'Lucro']
df_mes = df_2019.groupby('Mes')[colunas].sum().round(2)
df_mes['Margem_lucro'] = df_mes.Lucro / df_mes.Valor_Total_Desconto
df_mes['Margem_lucro'] = (df_mes.Lucro / df_mes.Valor_Total_Desconto * 100).round(1)
df_mes
```

```python
#criando graficos com matplotlib
plt.figure(figsize=(14,7))
plt.bar(x=df_mes.index, height=df_mes['Margem_lucro'], color = 'green')
plt.title('Margem de Lucro por Mês')
plt.xlabel('Mes')
plt.ylabel('% Margemde de lucro')
```

```python
meses_ordem = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
df_mes = df_mes.reindex(meses_ordem)

plt.figure(figsize=(20,3))

plt.plot(df_mes.index, df_mes.Valor_Total_Desconto, marker='o', color = 'orange')
plt.plot(df_mes.index, df_mes.Custo_Total, marker='o', linestyle='--', color = 'green')
plt.title('Margem de Lucro por Mês')
plt.xlabel('Mês')
plt.ylabel('% Margemde de Lucro')
plt.grid(True)
```
