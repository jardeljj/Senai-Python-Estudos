## Gráfico de linhas

 ### Matplotlib
 - [Documentação gráficos de linha](https://matplotlib.org/2.0.2/api/lines_api.html)

```python
df.head()
```

```python
df['sepal_length'].plot()
```

```python
df['sepal_length'].plot(figsize=(15, 5), 
                        linewidth=2,
                        color='green',
                        title='Tamanha da Sépala',
                        xlabel='Índice',
                        ylabel = 'tamanho em cm',
                        marker='o',
                        linestyle='-.',
                        legend=True)
```

```python
plt.figure(figsize=(15, 5)) # configura o tamanho da área de plotagem

plt.plot(df.index, df['sepal_length'])
plt.plot(df.index, df['sepal_width'], color='gray', linewidth=1, linestyle='--')
plt.title('Sépala')
plt.xlabel('ìndice')
plt.ylabel('Tamanho em cm'),
plt.legend(['altura', 'largura'])
plt.show()
```

### Subplots

```python
df.head()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(df.index, df['sepal_length'], color='green')
plt.title('Altura da sépala')
plt.ylabel('tamanho em cm')

plt.subplot(1, 3, 2)
plt.plot(df.index, df['sepal_width'], color='red')
plt.title('Largura da sépala')

plt.subplot(1, 3, 3)
plt.plot(df.index, df['petal_length'], color='yellow')
plt.title('Altura da pétala')

plt.show()
```

## Gráfico de barras

### Matplotlib

 - [Documentação gráfico de barras verticais](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html)
 - [Documentação gráfico de horizontais](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.barh.html)

```python
df.head()
```

```python
df['sepal_length'].head().plot(kind='bar')
```

```python
df['sepal_length'].head()
```

```python
# contando valores e plotando um gráfico de barras
df['species'].value_counts().plot(kind='bar', 
                                  title='Contagem de espécies',
                                  color='green')
```

```python
# utilizando o Matplotlib

plt.figure(figsize=(15,5))
plt.bar(x=df['species'], height=df['species'].count())
plt.title('Contagem de espécies')
plt.xlabel('Espécie')
plt.ylabel('Contagem')
```

```python
# barras horizontais
plt.figure(figsize=(15,5))
plt.barh(width=df['species'].count(), y=df['species'])
plt.title('Contagem de espécies')
plt.xlabel('Contagem')
plt.ylabel('Espécie')
```

### Seaborn (gráfico de barras)

- [Documentação gráfico de barras](http://seaborn.pydata.org/generated/seaborn.barplot.html)

```python
import numpy as np 

df_agrupado =  df.groupby('species').agg({'sepal_length': np.median})
type(df_agrupado)
```

```python
df_agrupado
```

```python
# barras verticais
sns.barplot(data=df_agrupado, x=df_agrupado.index, y='sepal_length')
```

```python
# barras horizontais
sns.barplot(data=df_agrupado, x='sepal_length', y=df_agrupado.index)
plt.title('Barras horizontais')
```

## Gráfico de Pizza

### Matplotlib
- [Documentação gráfico pizza](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html)

```python
df.head()
```

```python
df_contagem_especies = df['species'].value_counts()
```

## Nuvem de palavras (Wordcloud)

- `pip install wordcloud`
- [Documentação](https://github.com/amueller/word_cloud)

```python
!pip install wordcloud
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

### Lendo o arquivo TXT

```python
with open('wordcloud/elefante.txt', 'r', encoding='utf8') as file:
    texto = file.read()
```

```python
print(texto)
```

```python
wc = WordCloud(height=400, width=400)
```

```python
type(wc)
```

```python
nuvem = wc.generate(texto)
```

### Plotando a nuvem de palavras

```python
plt.imshow(nuvem)
plt.axis('off')
plt.show()
```

### Stopwords

```python
stopwords = ['o', 'a', 'um', 'uma', 'â', 'e', 'ela', 'ele', 'não', 'mas', 'que', 'por', 'ali']
```

```python
wc = WordCloud(height=400, width=400, stopwords=stopwords)
```

```python
nuvem = wc.generate(texto)
plt.imshow(nuvem)
plt.axis('off')
plt.show()
```

### Criando uma máscara



```python
from PIL import Image
import numpy as np
```

```python
imagem = Image.open('wordcloud/elefante.png')
```

```python
plt.imshow(imagem)
plt.axis('off')
plt.show()
```

```python
mascara = np.array(imagem)
```

```python
type(mascara)
```

```python
mascara.shape
```

```python
wc = WordCloud(mask=mascara, 
               background_color='white', 
               stopwords=stopwords,
               contour_color='black',
               contour_width=1)
```

```python
nuvem = wc.generate(texto)
```

```python
plt.imshow(nuvem)
plt.axis('off')
plt.show()
```
